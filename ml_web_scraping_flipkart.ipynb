```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as ureq
url='https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
url_req=ureq(url)
html_page=url_req.read()
url_req.close()
page_soup=soup(html_page,'html.parser')
details=page_soup.findAll('div',{"class":"_3O0U0u"})
print(len(details)) # gives the no of product on the first page
```

```python
filename='iphone_detail.csv'
f=open(filename,'w')
headers='Product_name,Pricing,Ratings\n'
f.write(headers)
for model in details:
    product_name=model.div.img['alt']
    model_price=model.findAll('div',{"class":"col col-5-12 _2o7WAb"})
    price=model_price[0].text.strip()
    model_rating=model.findAll('div',{"class":"niH0FQ"})
    ratings=model_rating[0].text
    # string parsing
    trim_price=''.join(price.split())
    rupee=trim_price.split('â‚¹')
    rs_rupee='Rs.'+rupee[1]
    #print(rs_rupee)
    split_price=rs_rupee.split('E')
    final_price=split_price[0]
    final_price=final_price.replace(',','')
    split_rating=ratings.split(" ")
    split2_rating=split_rating[0].split(",")
    a=split2_rating[0]
    final_rating="{0:0.1f}".format(float(a))
    #print('Product_name:'+ product_name)
    #print('Pricing:'+ final_price)
    #print('Ratings:'+final_rating)
    print(product_name.replace(",",'|')+','+ final_price+','+final_rating+'\n')
    f.write(product_name.replace(",",'|')+','+ final_price+','+final_rating+'\n')
f.close()
```
