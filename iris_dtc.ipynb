```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
iris=load_iris()
print(iris.keys())
print('feature names are:\n',iris.feature_names)
print('what to predict:',iris.target_names)
```

```python
iris_ds=pd.DataFrame(iris.data)
iris_ds.columns=iris.feature_names
label=pd.DataFrame(iris.target,columns=['type of flower'])
iris_comp=iris_ds.join(label['type of flower'])
iris_comp.head()
```

```python
cols=iris_ds.columns
def scatter_plot(cols):
    for col in cols:
        x_ax=iris_ds[col]
        y_ax= label
        plt.figure(figsize=(5,4))
        plt.scatter(x_ax,y_ax, alpha=0.4)
        plt.title(col+' vs '+ 'type of flower')
        plt.xlabel(col)
        plt.ylabel('type of flower')
scatter_plot(cols)
```

```python
sns.pairplot(iris_comp)
```

```python
x=iris.data
y=iris.target
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=.30,random_state=40)
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
pred=model.predict(xtest)
print('score of training is:',model.score(xtrain,ytrain))
print('accuracy score is:',accuracy_score(ytest,pred))
print('confusion matrix is: \n',confusion_matrix(ytest,pred))
print('classification report is :\n', classification_report(ytest,pred))
print('0-Setosa \n1-Versicolor \n2-Virginica')
predict=pd.DataFrame(pred)
predict.columns=['Predicted values']
predict
```
